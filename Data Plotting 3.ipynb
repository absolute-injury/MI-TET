```python
import pandas as pd
import matplotlib.pyplot as plt

files = {
    'default': 'resuls_default.csv',
    'shallow': 'resuls_shallow.csv',
    'deep_BP': 'resuls_deep_BP.csv'
}
files_ = {
    'default': 'results_default.csv',
    'shallow': 'results_shallow.csv',
    'deep_BP': 'results_deep_BP.csv'
}

colors = {'default': 'indigo', 'shallow': 'teal', 'deep_BP': 'yellowgreen'}

plt.figure(figsize=(12, 8))

for name, filepath in files.items() :
    df = pd.read_csv(filepath)

    rewards = df['rewards']

    window = 10
    rolling_mean = rewards.rolling(window=window).mean()
    rolling_std = rewards.rolling(window=window).std()

    plt.scatter(rewards.index, rewards, color=colors[name], s=10, alpha=0.15)
    plt.plot(rewards.index, rolling_mean, color=colors[name], linewidth=2, label=name)
    plt.fill_between(
        rewards.index,
        rolling_mean - rolling_std,
        rolling_mean + rolling_std,
        color=colors[name], alpha=0.2
    )

plt.xlabel('Episode', fontsize=16)
plt.ylabel('Reward', fontsize=16)
plt.legend(loc='upper left', fontsize=18)
plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.savefig("learning_curve_B50.png", dpi=300)
```
